## Array Arithmetics

```python
my_list_1 = [1,2,3]
my_list_2 = [4,5,6]
```

```python
my_list_1 + my_list_2
```

```python
my_array_1 = np.array(my_list_1)
my_array_2 = np.array(my_list_2)
```

```python
my_array_1 + my_array_2
```

```python
my_array_1 + np.ones(my_array_1.shape)
```

```python
my_array_1 - my_array_2
```

```python
my_array_1 * my_array_2
```

```python
my_array_1 / my_array_2
```

```python
my_array_1 * 10
```

```python
my_array_1.min()
```

```python
my_array_2.max()
```

```python
my_array_2.shape
```
