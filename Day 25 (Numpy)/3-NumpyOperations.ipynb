### z-score

```python
data = np.array([10,12,13,15,18,25,100,105])
```

### outlier

```python
mean = data.mean()
```

```python
mean
```

```python
data[data > mean]
```

```python
std = data.std()
```

```python
std
```

```python
z_scores = (data - mean) / std
```

```python
z_scores
```

```python
data[z_scores > 1]
```
